```python
import pandas as pd
import numpy as np

# Create synthetic dataset
np.random.seed(42)
data = pd.DataFrame({
    'area': np.random.randint(500, 3000, 100),
    'bedrooms': np.random.randint(1, 5, 100),
    'bathrooms': np.random.randint(1, 3, 100),
})

# price formula (custom logic)
data['price'] = data['area']*300 + data['bedrooms']*50000 + data['bathrooms']*30000 + np.random.randint(-20000,20000,100)

data.head()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = data[['area','bedrooms','bathrooms']]
y = data['price']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = LinearRegression()
model.fit(X_train,y_train)

print("Accuracy:", model.score(X_test,y_test))
model.predict([[2000,3,2]])
```
